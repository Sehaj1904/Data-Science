## **Realme mobile**

```python
flipkart = requests.get("https://www.flipkart.com/search?q=realme+mobile")
```

```python
soup = BeautifulSoup(flipkart.text, "html.parser") # hum apne code ko sahi banane ke liye html me kiya hai 
```

```python
print(soup.prettify) # prettify bss code ko structured form me likhta hai
```

```python
soup.title
```

```python
soup.title.text # isme aur upar vale bss ye farak hai ki isme <title> likha hua nhi aa raha just bcoz .text use kiya hai
```

```python
soup.p # .p just 1st para ka content deta hai
```

```python
soup.div.text # .div.text 1st para ka poora detail dega
```

```python
mobile = soup.find("div", class_="_1AtVbE") # .find bss ek ko find kardega  
```

```python
mobile
```

```python
# we can write by this way class_="_1AtVbE" and also by this way attrs = {"class": "_1AtVbE"}
mobile = soup.find_all("div", attrs = {"class": "_1AtVbE"})
```

```python
# ye 32 galat hai as uss page me 24 ka data hai kyuki _1AtVbE ye class agar tum dekho realme page pe inspect me jaha ye class thi to dikhega 
# ki ye aur bhi faltu ki cheeze le raha hai iss class ke andar jiske karn 32 aaya to hume bss kitne phones hai uss page pe vo cahie  
len(mobile)
```

```python
# _13oc-S sahi class hai isliye len sahi diya hai 
mobiles = soup.find_all("div", attrs = {"class": "_13oc-S"}) # .find_all poore data ko consider karega jitne ek page me honge 
```

```python
len(mobiles)
```

```python
mobile = mobiles[0] # YE ISLIYE LIKHA HAI KYUKI humne dekha tha ki pehle find kiya tha to vo mobile ho bhi sakhta hai aur nhi kyuki mobile 
# ki len 32 thi jo galat thi to hum mobiles[0] store kar rahe hai mobile me taki pehle mobile ka name print ho jaye
name = mobile.find("div", attrs = {"class": "_4rR01T"})
```

```python
name.text # name of ist moblile of the page
```

```python
price = mobile.find("div", class_= "_30jeq3 _1_WHN1")
```

```python
price.text  # price of 1st mobile 
```

```python
properties = mobile.find_all("li")  # li gives all properties of 1st mobile that are written below the name of the mobile
```

```python
properties
```

```python
len(properties)
```

```python
ram = properties[0] # ye hume saari properties me se pehli property dega 
```

```python
ram 
```

```python
ram.text
```

```python
# pehle humne | jaha jaha hai vaha vah split kiya hai aur output list form me mil gaya dekh lo
ram.text.split(" | ") 
```

```python
ram.text.split(" | ")[0] # bss ram ki value cahie to aise aati hai aur ye hume array ka pehla element dega in short as index 0 hai
```

```python
names = []  # yaha hum saare mobiles ka data fetch kar rahe hai mobiles se aur mobile keechega data 
prices = [] # dataframe me store karn ke liye aisa likha hai (kahi pe store kiya hai)
rams = []
for mobile in mobiles:
  name = mobile.find("div", attrs = {"class": "_4rR01T"}).text
  price = mobile.find("div", class_= "_30jeq3 _1_WHN1").text
  ram = mobile.find_all("li")[0].text.split("|")[0]  # humne saari properties me se bss ram vali fetch ki hai 
  names.append(name)
  prices.append(price)
  rams.append(ram)
```

```python
names # jaise names aaya hai vaise hi prices and rams bhi aa jaegi 
```

```python
dic = {"Model" : names, "Price" : price, "RAM" : ram} # dictionary isliye taki hum data frame me convert kar sake 
```

```python
import pandas as pd
df = pd.DataFrame(dic)
```
