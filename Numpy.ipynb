```python
lst = [1,2,3,4,5] # ye list hai
lst2 = lst[2:4]
print(lst) # lst and lst 2 are different as when we change lst2 but it dosen't affect lst 
print(lst2) # and we see that lst2 lst se hi bann raha hai but still dono alag vo 
lst2[0] = 1000 # lst2[0] vale se pata chal gaya
print(lst)
print(lst2)
```

```python
arr = np.array([1,2,3,4,5]) # ye array hai 
arr2 = arr[2:4]
print(arr)
print(arr2)
arr2[0] = 1000
print(arr)
print(arr2)
```

```python
arr = np.array([1,2,3,4,5])
arr2 = arr[2:4].copy()
print(arr)
print(arr2)
arr2[0] = 1000
print(arr)
print(arr2)
```
