```python
import seaborn as sns
```

```python
# tip target variable hai aur dependent hai aur issi ki prediction karni hai bakiyo se relation nikal kebaki independent hai 
df = sns.load_dataset('tips') 
```

```python
df
```

```python
sns.relplot(x='total_bill',y='tip',data=df)
```

```python
# hue differentiate karta hai bascically jaise sex me male ya female
# size bss differentiate karta hai mtlb chota usse bada use bada circle jaise graph me dekho 
sns.relplot(x='total_bill',y='tip',data=df,hue='sex',style='time',size='size')
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='sex',style='time',size='size',kind='line')
```

```python
sns.lineplot(x='total_bill',y='tip',data=df)
```

```python
# col bss do alag alag banake de raha hai time ko jaise lunch dinner ka alag alag 
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker',col='time')
```

```python
# jitter=False ek line me plot karta hai data ko overlap karta hai paralley dikhata hai  scattered nhi
sns.catplot(x='day',y='total_bill',data=df,jitter=False) 
```

```python
# swarm bss outliers ko dekhne ke liye hota hai
sns.catplot(x='day',y='total_bill',data=df,kind='swarm',hue='size')
```

```python
sns.catplot(x='smoker',y='tip',data=df,order=['No','Yes'])
```

```python
# boxplot 0 25(quater1) 50(median) 75(quater3) 100 vala dikhata hai 
sns.catplot(x='day',y='tip',data=df,kind='box')
```

```python
sns.catplot(x='day',y='tip',data=df,kind='box',hue='sex')
```

```python
sns.catplot(x='total_bill',y='day',data=df,hue='sex',kind='violin')
```

```python
# split=True isliye lika kyuki hume ek me hi dono cahie the agar split na likhte to ye har ek ko 2 plot banata ek blue and orange 
# alag alag har ek ka but hume ek hi vilion plot ko split kiya 2 me taki alag alag na aaye Compare upar and ye vala 

sns.catplot(x='total_bill',y='day',data=df,hue='sex',kind='violin',split=True)

# OR 
# isme humne split ko string form me liya hai aur string form me hamesha true hi hota hai cahie hum false bhi likhe kyuki 
# vo false string form me hai

# sns.catplot(x='total_bill',y='day',data=df,hue='sex',kind='violin',split='false')
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='violin',inner=None)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df,color='k',size=3)
```

```python
g = sns.catplot(x='day',y='total_bill',data=df,kind='violin',inner=None)
sns.swarmplot(x='day',y='total_bill',data=df,color='k',size=3,ax=g.ax)
```

```python
#pairplot only show numerical plot and it gives all possible relations ka plot   (This shows the relationship for (n, 2) combination of 
# variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots.)
sns.pairplot(df)
```

```python
# show realtion b/w numerical plot only not incase of categorial plot
df.corr()
```
