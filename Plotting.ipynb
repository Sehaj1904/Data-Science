## **IMP TERMS**
The **plot()** function is used to draw markers on the graph but by default plot() draws lines from point to point.

Link for markers colors linestyles format strings :

*   ms = marker size
*   mfc = marker’s face color
*   mec = marker’s edge color

### Setting Title and Labels 

*   .title(“title name”)
*   .ylabel(“label”)
*   .xlabel(“label”)

 marker :  to emphasize each point with a specified marker.

 linestyle : to change the style of the plotted line.

 


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3]  
y = [4,5,6]

# we use plt.plot(x, y) to create our plotlines
# and we use plt.show() to make it visible to us

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,marker = 'o')
plt.show()
```

```python
plt.plot(x,y,marker = 'o',linestyle = ' ')
plt.show()
```

```python
plt.plot(x,y,marker = 'D',linestyle = 'dashed',color ='red')
plt.show()
```

```python
plt.plot(x,y,marker='1',linestyle='dashed',ms=30,mfc='red',mec='black')
```

```python
plt.plot(x,y,marker='o',linestyle='dashed',ms=30,mfc='red',mec='black')
plt.show()
```

```python
plt.plot(x,y,marker='o',linestyle='dashed',ms=30,mfc='red',mec='black',linewidth = 3)
plt.show()
```

# **Scatter**
**Size** denotes the size of scatter points on the graph, and **alpha** resembles opacity.

```python
plt.scatter([3,9,7],[100,200,89])
plt.show()
```

```python
plt.scatter([3,9,7],[100,200,89])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
plt.show()
```

```python
x = np.arange(5,25)
x 
```

```python
plt.scatter(x,x**2,color='r',label='x^2',marker='^')
plt.scatter(x,x**3,color='g',label='x^3',marker='o')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
plt.legend()    # gives the scale(label) of the graph
plt.show()
```

```python
plt.scatter(x,x**2,color='r',label='x^2',marker='^')
plt.scatter(x,x**3,color='g',label='x^3',marker='o')
plt.scatter(x,x**4,color='g',label='x^4',marker='o')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
plt.legend()
plt.grid()  # ye jo lines dikh rahi hai na vo grid se aati hai (to add grid lines to the plot) 
plt.show()
```

```python
x = np.round(np.random.rand(7)*7)
x
```

```python
y = np.round(np.random.rand(7)*10)
y 
```

```python
plt.scatter(x,y,s=[10000,1200,1300,400,1300,1200,1100],alpha=0.3,color=['r','b','g','y','g','c','g']) # 
plt.plot()  # s is size of the scatter points and alpha resembles opacity (transparency : jitna kam utna light jitna jada utna dark)
```

# **SubPlots**

```python
plt.subplot(1,2,1)
plt.plot([1,2,3],[7,8,9])
plt.title('Pepcoding')
plt.subplot(1,2,2)
plt.plot([4,5,6],[7,8,9])
plt.title('Nados')
plt.show()
```

```python
plt.subplot(2,3, 1)

x = np.round(np.random.rand(7)*7)
y = np.round(np.random.rand(7)*9)

plt.scatter(x, y, s=[1120, 1100, 1500, 1300, 1500, 1800, 2000], alpha=0.4, color = ['r', 'g', 'b', 'y', 'r', 'b', 'b'])


plt.subplot(2,3, 3)
x = np.arange(2, 20)
plt.scatter(x, x**2, color = 'r', label='x^2', marker = '^')
plt.scatter(x, x**2.5, color = 'g', label='x^2.5', marker = 'o')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('x * y')
plt.legend()

plt.subplot(2,3, 5)
plt.scatter([3, 7, 5], [2, 4, 8])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter')

plt.subplot(2,3, 4)
plt.scatter([3, 7, 5], [2, 4, 8])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter')


plt.show()
```

# **Bar Chart**

```python
plt.bar(['Jan','Feb','Mar','Apr'],[10,20,6,22],color=['r','g','b','y'],width=[0.4,0.6,1,0.9])

plt.xlabel('Months')
plt.ylabel('Crime')
```

```python
plt.barh(['Jan','Feb','Mar','Apr'],[10,20,6,22],color=['r','g','b','y'],height=[0.4,0.6,1,0.9])

plt.xlabel('Months')
plt.ylabel('Crime')
```
