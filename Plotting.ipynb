```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3]  
y = [4,5,6]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,marker = 'o')
plt.show()
```

```python
plt.plot(x,y,marker = 'o',linestyle = ' ')
plt.show()
```

```python
plt.plot(x,y,marker = 'D',linestyle = 'dashed',color ='red')
plt.show()
```

```python
plt.plot(x,y,marker='1',linestyle='dashed',ms=30,mfc='red',mec='black')
```

```python
plt.plot(x,y,marker='o',linestyle='dashed',ms=30,mfc='red',mec='black')
plt.show()
```

```python
plt.plot(x,y,marker='o',linestyle='dashed',ms=30,mfc='red',mec='black',linewidth=3)
plt.show()
```

# **Scatter**

```python
plt.scatter([3,9,7],[100,200,89])
plt.show()
```

```python
plt.scatter([3,9,7],[100,200,89])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
plt.show()
```

```python
x = np.arange(5,25)
x 
```

```python
plt.scatter(x,x**2,color='r',label='x^2',marker='^')
plt.scatter(x,x**3,color='g',label='x^3',marker='o')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
plt.scatter(x,x**2,color='r',label='x^2',marker='^')
plt.scatter(x,x**3,color='g',label='x^3',marker='o')
plt.scatter(x,x**4,color='g',label='x^4',marker='o')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
plt.legend()
plt.grid()
plt.show()
```
